```python
from keras.datasets import mnist
import matplotlib.pyplot as plt
import numpy as np
```

```python
data = mnist.load_data()
```

```python
data
```

```python
((X_train, y_train), (X_test, y_test)) = data 
```

```python
X_shape = X_train.shape[0]
X_shape1 = X_test.shape[0]
```

```python
X_train = X_train.reshape((X_shape, 28*28)).astype('float32') 
X_test = X_test.reshape((X_shape1, 28*28)).astype('float32')
```

```python
print(X_train[0])
print()
print(X_test[0])
```

```python
X_train = X_train/255
X_test = X_test/255
```

```python
from keras.utils import np_utils
```

```python
y_train = np_utils.to_categorical(y_train)
y_test = np_utils.to_categorical(y_test)

#print(y_test.shape)
num_train_class = y_train.shape[1]
print(num_train_class)

num_test_class = y_test.shape[1]
print(num_test_class)
```

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()

model.add(Dense(32,  input_dim = 28*28, activation = 'relu'))

model.add(Dense(64, activation = 'relu'))

model.add(Dense(10, activation = 'softmax'))
```

```python
model.compile(loss = 'categorical_crossentropy', optimizer = 'adam', metrics = ['accuracy'])
```

```python
model.summary()
```

```python
model.fit(X_train, y_train, epochs = 10, batch_size = 100) 
```

```python
scores = model.evaluate(X_test, y_test)
print(scores)
```
